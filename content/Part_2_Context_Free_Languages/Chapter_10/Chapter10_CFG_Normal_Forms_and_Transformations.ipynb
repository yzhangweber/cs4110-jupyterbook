# Normal Forms and Transformations in Context-Free Grammars

## Introduction

This chapter explores various normal forms and transformations in context-free grammars (CFGs). We will discuss algorithms to convert grammars into different normal forms and demonstrate their properties through examples.


## 1. Regular Grammar

### 1.1 Regular Grammar Definition:

A regular grammar is a special type of context-free grammar where production rules follow specific patterns. In a regular grammar, all productions must be in one of these forms:
- $A → xB$ (where A, B are single non-terminals and x is a string of terminals, x can be empty)
- $A → x$ (where A is a single non-terminal and x is a string of terminals, x can be empty)

### 1.2 Regular Grammar Examples:

#### Example 1: Regular Grammar
Given this Grammar:
- V: {S, A}
- $\Sigma$: {a, b}
- R: $\{S → aA | bB | a; A → aA | bA | ∧\}$

This is a regular grammar because each production follows the specific patterns outlined for a regular grammar.

#### Example 2: Regular Grammar
Given this Grammar:
- V: {S, A, B}
- $\Sigma$: {a, b}
- R: $\{S → A | ∧; A → aB; B → aA | bB | b\}$

This is a regular grammar because each production follows the specific patterns outlined for a regular grammar.

### 1.3 Regular Grammar to Regular Expression:
Regular grammars are equivalent in power to regular expressions. Any language that can be described by a regular expression can also be described by a regular grammar, and vice versa. 

#### Example 3: Regular Grammar to Regular Expression
Given the regular grammar:
- V: {S}
- $\Sigma$: {a, b}
- R: $\{S → aS | b\}$

This CFG corresponds to the regular expression **aa*b**, because the **aa\*** means one or more occurrences of 'a' and is followed by 'b'.

#### Example 4: Regular Grammar to Regular Expression
Given the regular grammar:
- V: {S}
- $\Sigma$: {a, b}
- R: $\{S → abS | baS | ∧\}$

This is a regular grammar because each production follows the specific patterns outlined for a regular grammar. We can observe that every string generated by this grammar will consist of any number of occurrences of the substrings "ab" and "ba" concatenated in some order, and the production S → ∧ allows us to stop the derivation and terminate with the empty string. The language generated by this grammar consists of strings that can be a finite sequence of the substrings "ab" and "ba" (the strings consist of any number of "ab"s and "ba"s in any order). Thus, it can be expressed by the regular expression: **(ab + ba)\***.

### 1.4 Finite Automata to Regular Grammar:
Regular grammars have a direct correspondence with finite automata. In a regular grammar, non-terminals correspond to states in the automaton, productions correspond to transitions, terminal productions correspond to accepting states, and the start symbol corresponds to the initial state. ### 1.5 Code Example: Representing a CFG in Python

The formal definition of a CFG is G = (V, Σ, R) with start symbol S. Before we can implement any transformation algorithm (null removal, unit removal, CNF conversion), we need a concrete data structure to hold these components. This cell defines two classes used by **all subsequent code cells** in this chapter:

| Class | Represents | Key attributes |
|---|---|---|
| `Production` | A single rule A → α | `left` (str), `right` (list of str) |
| `Grammar` | The full CFG (V, Σ, R, S) | `productions`, `start_symbol`, `terminals`, `non_terminals` |

**Design convention used throughout this chapter:**
- Non-terminals are uppercase single letters: `'S'`, `'A'`, `'B'`, …
- Terminals are lowercase letters or non-alpha symbols: `'a'`, `'b'`, `'∧'`, …
- The empty string is represented as the single-character string `'∧'`

The `Grammar` class infers V (non-terminals) and Σ (terminals) automatically
from the production rules, using the case convention above. This means you
never need to pass V and Σ explicitly — just provide the productions and start
symbol.

**Run this cell first.** Every other code cell in the chapter imports
`Production` and `Grammar` from this cell's scope.


````python
# ─────────────────────────────────────────────────────────────────────────────
# Core Data Structures: Production and Grammar
#
# These two classes are used by every transformation function in this chapter.
# Run this cell before any other code cell.
#
# Running outside Jupyter:
#   Include this code at the top of any .py file that uses the functions below,
#   or place it in a separate file (e.g., cfg_core.py) and import it.
# ─────────────────────────────────────────────────────────────────────────────


class Production:
    """
    Represents a single CFG production rule of the form A → α.

    Attributes:
        left  (str):       The non-terminal A on the left-hand side.
        right (list[str]): The string α on the right-hand side, stored as a
                           list of symbol strings (e.g., ['a', 'A'] for aA).
    """
    def __init__(self, left, right):
        self.left = left    # LHS: always a single non-terminal (uppercase)
        self.right = right  # RHS: list of terminals and/or non-terminals

    def __str__(self):
        # Pretty-print as "A → aB" for display
        return f"{self.left} → {''.join(self.right)}"


class Grammar:
    """
    Represents a full CFG G = (V, Σ, R) with start symbol S.

    V (non_terminals) and Σ (terminals) are inferred automatically from the
    production rules using the convention that non-terminals are uppercase and
    terminals are lowercase or non-alphabetic.

    Attributes:
        productions   (list[Production]): All production rules R.
        start_symbol  (str):              The start symbol S.
        terminals     (set[str]):         Σ — inferred from productions.
        non_terminals (set[str]):         V — inferred from productions.
    """
    def __init__(self, productions, start_symbol):
        self.productions  = productions
        self.start_symbol = start_symbol
        self.terminals     = self._find_terminals()
        self.non_terminals = self._find_non_terminals()

    def _find_terminals(self):
        """
        Collect Σ: any symbol on a RHS that is lowercase or non-alphabetic.
        '∧' (the empty string marker) is treated as a terminal here so that
        null productions are detected by the transformation functions.
        """
        terminals = set()
        for prod in self.productions:
            for symbol in prod.right:
                if symbol.islower() or not symbol.isalpha():
                    terminals.add(symbol)
        return terminals

    def _find_non_terminals(self):
        """
        Collect V: the start symbol plus any uppercase symbol appearing on
        either side of any production rule.
        """
        non_terminals = {self.start_symbol}
        for prod in self.productions:
            non_terminals.add(prod.left)          # LHS is always a non-terminal
            for symbol in prod.right:
                if symbol.isupper():              # Uppercase RHS symbols are non-terminals
                    non_terminals.add(symbol)
        return non_terminals

    def __str__(self):
        return '\n'.join(str(prod) for prod in self.productions)
````

## 2. Removing Null Productions

### 2.1 What Are Null Productions?
A null production (also called ∧-production) is a production of the form: A → ∧, where A is a non-terminal, and ∧ represents the empty string.

### 2.2 Why Remove Null Productions?
While these productions can be convenient for describing certain language features, removing them can simplify grammar analysis and parsing.

* Simplified Parsing: Many parsing algorithms (like CYK) work better without null productions
* Easier Analysis: Grammar properties become easier to analyze without null productions
* Prerequisite for Normal Forms: Converting to Chomsky Normal Form or Greibach Normal Form requires eliminating null productions

### 2.3 Important Properties
* Removing null productions preserves the language defined by the grammar. The only exception is if the original language includes the empty string as a valid word; in that case, the empty string will be omitted from the new grammar, and we can handle this special case separately.
* The process of removing null productions may increase the number of productions in the grammar
* The resulting grammar generates exactly the same non-empty strings as the original grammar

### 2.4 Steps to Remove Null Productions
Follow these steps to remove null productions from a context-free grammar (CFG):
1. Find Nullable Non-terminals:
    - Mark all non-terminals A's with direct productions A → ∧
    - Iteratively mark any non-terminal A whose right-hand side consists entirely of nullable symbols, such non-terminal has indirect null productions
    - Continue until no more non-terminals can be marked nullable
2. Generate New Productions:
    - For each production A → X₁X₂...Xₙ
    - Create new productions by omitting nullable symbols in all possible combinations
    - Exclude any original and resulting null productions

### 2.5 Examples to Remove Null Productions
These following examples demonstrate various scenarios you might encounter when removing null productions, from simple cases to more complex ones involving multiple nullable symbols and nested dependencies. 

#### Example 1: Grammar with One Nullable Non-terminal
Consider this grammar:
```
S → AB
A → aA | ∧
B → b
```
Step 1: Find nullable non-terminals: 
- A → ∧ directly, so A is nullable.
- No other nullable symbols found.  

Step 2: Generate new productions: From S → AB, since A is nullable, generate the following new productions:  
- S → B (removing A from AB since A is nullable) 
- A → a (removing A from aA since A is nullable)
Keep all other non-null productions.

Final grammar:
```
S → AB | B
A → aA | a
B → b
```

#### Example 2: Grammar with Multiple Nullable Non-terminals
Consider this grammar:
```
S → ABC
A → aA | ∧
B → bB | ∧
C → c
```
Step 1: Find nullable non-terminals: 
- A → ∧ directly, so A is nullable.
- B → ∧ directly, so B is nullable.
   
Step 2: Generate new productions: 
* From S → ABC, generate all combinations with nullable A and B:
  - S → BC (removing A from ABC since A is nullable) 
  - S → AC (removing B from ABC since B is nullable)
  - S → C (removing both A and B from ABC since A and B are nullable)
* From A → aA | ∧, generate all combinations with nullable A:
  - A → a (removing A from aA since A is nullable) 
* From B → bB | ∧, generate all combinations with nullable B:
  - B → b (removing B from bB since B is nullable) 
* Keep all other non-null productions.

Final grammar:
```
S → ABC | BC | AC | C
A → aA | a
B → bB | b
C → c
```

#### Example 3: Complex Grammar with Chain of Nullable Non-terminals
Consider this grammar:
```
S → AB
A → CD
B → b
C → c | ∧
D → d | ∧
```
Step 1: Find nullable non-terminals: 
- C → ∧ directly, so C is nullable.
- D → ∧ directly, so D is nullable.
- Check combinations: since CD can be null, confirming A is nullable, in other words, A can indirectly produce an empty string
   
Step 2: Generate new productions: 
* From A → CD, generate all combinations with nullable C and D:
  - A → C (removing D from CD since D is nullable) 
  - A → D (removing C from CD since C is nullable)
* From S → AB, generate all combinations with nullable A:
  - S → B (removing A from AB since A is nullable) 
* Keep all other non-null productions.

Final grammar:
```
S → AB | B
A → CD | C | D
B → b
C → c
D → d
```

### 2.6 Code Example: Removing Null Productions

The algorithm below implements the two-step procedure from Section 2.4 exactly as described. Before reading the code, recall the key insight: a non-terminal is **nullable** if it can eventually derive the empty string: either directly via A → ∧, or indirectly because every symbol on some RHS is itself nullable.

The implementation uses a **fixed-point iteration**: it keeps scanning all productions and marking new nullable non-terminals until a full pass finds nothing new. This is the standard way to compute the transitive closure of "can derive ∧".

**How the two functions relate:**
* find_nullable_symbols(grammar) returns: set of nullable non-terminal strings
* remove_null_productions(grammar)
    * calls find_nullable_symbols to get nullable set
    * generates all subsets of nullable positions for each production
    * returns: a new Grammar with null productions eliminated

The `itertools.combinations` call inside `remove_null_productions` is doing the combinatorial work from Step 2 of Section 2.4: for a production like S → ABC where A and B are nullable, it generates all subsets of {A, B} to omit — giving S → ABC, S → BC, S → AC, S → C.

**Special case:** If the start symbol S is itself nullable (i.e., the language contains the empty string), the algorithm introduces a new start symbol S' with S' → S | ∧. This preserves membership of ∧ in the language while satisfying the requirement that the new grammar has no null productions except possibly at the new start symbol.

**Outside Jupyter:** Copy both functions and the `Grammar`/`Production` classes into a single `.py` file and run with `python null_removal.py`.

````python
# ─────────────────────────────────────────────────────────────────────────────
# Step 1 of the transformation pipeline: Remove Null Productions
#
# Implements Section 2.4 exactly:
#   Phase 1 → find_nullable_symbols (fixed-point iteration)
#   Phase 2 → remove_null_productions (combinatorial expansion)
#
# Running outside Jupyter:
#   python null_removal.py
# ─────────────────────────────────────────────────────────────────────────────

from itertools import combinations  # used to generate all subsets of nullable positions


def find_nullable_symbols(grammar):
    """
    Find all nullable non-terminals: those that can derive the empty string ∧.

    Uses fixed-point iteration (Section 2.4, Step 1):
      - Direct nullable:   A → ∧
      - Indirect nullable: A → X₁X₂…Xₙ where every Xᵢ is already nullable

    Returns a set of nullable non-terminal strings.
    """
    nullable = set()
    changed = True  # loop until a full pass adds nothing new (fixed point)

    while changed:
        changed = False
        for prod in grammar.productions:
            if prod.left in nullable:
                continue  # already marked, skip

            # Case 1: direct null production A → ∧
            if len(prod.right) == 1 and prod.right[0] == '∧':
                nullable.add(prod.left)
                changed = True

            # Case 2: all symbols on the RHS are already known to be nullable
            # (indirect null production — e.g., A → CD where C and D are nullable)
            elif all(symbol in nullable for symbol in prod.right):
                nullable.add(prod.left)
                changed = True

    return nullable


def remove_null_productions(grammar):
    """
    Remove null productions while preserving the language (Section 2.4).

    For each production A → X₁X₂…Xₙ, generates every possible combination
    of omitting nullable symbols on the RHS. Empty resulting RHS strings are
    discarded (they would re-introduce null productions).

    Special case (Section 2.3): if the start symbol S is nullable, introduces
    a new start symbol S' → S | ∧ to preserve the empty string in the language.

    Returns a new Grammar without null productions (except S' → ∧ if needed).
    """
    # Phase 1: identify all nullable non-terminals
    nullable = find_nullable_symbols(grammar)
    print(f"Nullable symbols: {nullable}")

    # Phase 2: generate new productions
    new_productions = []
    for prod in grammar.productions:

        # Skip original null productions — they will be replaced by the
        # new productions we generate below
        if len(prod.right) == 1 and prod.right[0] == '∧':
            continue

        # Locate every position in the RHS that holds a nullable symbol
        nullable_positions = [
            i for i, symbol in enumerate(prod.right) if symbol in nullable
        ]

        # Generate all 2^k subsets of nullable positions to omit
        # (k = number of nullable positions in this production)
        for r in range(len(nullable_positions) + 1):
            for positions_to_remove in combinations(nullable_positions, r):
                # Build new RHS by keeping all symbols NOT in positions_to_remove
                new_right = [
                    sym for i, sym in enumerate(prod.right)
                    if i not in positions_to_remove
                ]
                # Only add if non-empty (never introduce a new null production)
                if new_right:
                    new_productions.append(Production(prod.left, new_right))

    # Special case: if S is nullable, add a new start symbol S' → S | ∧
    # so that the empty string is still in the language
    if grammar.start_symbol in nullable:
        new_start = f"{grammar.start_symbol}'"  # e.g., S → S'
        new_productions.append(Production(new_start, [grammar.start_symbol]))
        new_productions.append(Production(new_start, ['∧']))
        return Grammar(new_productions, new_start)

    return Grammar(new_productions, grammar.start_symbol)


def analyze_grammar_transformation(original, transformed):
    """
    Report the structural changes caused by a grammar transformation.
    Useful for verifying that null/unit removal behaves as expected.
    """
    original_prods   = len(original.productions)
    transformed_prods = len(transformed.productions)

    print(f"Number of productions: {original_prods} → {transformed_prods}")
    print(f"Production increase factor: {transformed_prods / original_prods:.2f}x")

    # Check whether any new non-terminals were introduced (e.g., S' for null start)
    new_nt = transformed.non_terminals - original.non_terminals
    if new_nt:
        print(f"New non-terminals added: {new_nt}")

    # '∧' disappearing from terminals is expected and correct:
    # it means all null productions have been successfully eliminated
    if original.terminals != transformed.terminals:
        print("Note: Terminal set changed (expected — '∧' should be gone).")
        print(f"  Original terminals:    {original.terminals}")
        print(f"  Transformed terminals: {transformed.terminals}")


# ── Example 1: one nullable non-terminal (matches Section 2.5 Example 1) ─────
simple_grammar = Grammar([
    Production('S', ['A', 'B']),
    Production('A', ['a', 'A']),
    Production('A', ['∧']),       # A → ∧  (direct null production)
    Production('B', ['b']),
], 'S')

print("Example 1: Grammar with one nullable non-terminal (A)")
print("Original grammar:")
print(simple_grammar)
print("\nAfter removing null productions:")
transformed = remove_null_productions(simple_grammar)
print(transformed)
print("\nTransformation analysis:")
analyze_grammar_transformation(simple_grammar, transformed)


# ── Example 2: multiple nullable non-terminals (matches Section 2.5 Example 2)
complex_grammar = Grammar([
    Production('S', ['A', 'B', 'C']),
    Production('A', ['a', 'A']),
    Production('A', ['∧']),       # A is directly nullable
    Production('B', ['b', 'B']),
    Production('B', ['∧']),       # B is directly nullable
    Production('C', ['c']),
], 'S')

print("\nExample 2: Grammar with multiple nullable non-terminals (A, B)")
print("Original grammar:")
print(complex_grammar)
print("\nAfter removing null productions:")
transformed = remove_null_productions(complex_grammar)
print(transformed)
print("\nTransformation analysis:")
analyze_grammar_transformation(complex_grammar, transformed)
````

## 3. Removing Unit Productions

### 3.1 What Are Unit Productions?
Unit productions (also called chain productions) are productions of the form A → B where both A and B are non-terminals. While these productions can make grammars more readable and intuitive, they do not add any meaningful structure to the derivation process and can make parsing less efficient. 

### 3.2 Why Remove Unit Productions?
1. Parsing Efficiency: unit productions can lead to unnecessary steps in parsing, creating longer derivation chains than needed.
2. Grammar Analysis: many algorithms (like CYK parsing) require grammars without unit productions.
3. Grammar Simplification: removing unit productions simplifies the grammar while preserving its language.

As an example, consider this grammar for arithmetic expressions:
```
E → T
T → F
F → var
F → num
F → (E)
```
While this grammar is readable, it requires three steps (E → T → F → var) to derive a simple variable. After removing unit productions, we get:
```
E → var
E → num
E → (E)
```
This transformed grammar requires a single step (E → var) to derive a simple variable.

### 3.3 How to Remove Unit Productions

#### 3.3.1 Unit Pairs and Transitive Closure
Before removing unit productions, we need to understand unit pairs and transitive closure. A unit pair (A,B) in a context-free grammar means that A can derive B through one or more (a sequence of) unit productions. The transitive closure of unit productions ensures that if there is a sequence of unit productions leading from one non-terminal to another, we establish a direct connection between them. For example, in the grammar:
```
S → A
A → B
B → C
C → a
```
Here are the steps to find all the unit pairs:
* Start with direct unit pairs: S → A, A → B, and B → C are direct unit pairs, so we add (S,A), (A,B), (B,C) to the unit pair set.
* Extend the pairs transitively: If (S,A) and (A,B) exist, it means S → B exists, then add (S,B) to the unit pair set. We repeat until no new pairs can be added. By extending the pairs transitively this way, We add more pairs (S,C), (A,C).

So the unit pairs are (S,A), (S,B), (S,C), (A,B), (A,C), (B,C).  

We can visualize unit pairs as a directed graph: solid arrows represent direct unit productions, and dashed arrows represent derived unit pairs.

```{mermaid}

    accDescr: a diagram representing unit pairs as a directed graph
    S[S] --> A[A]
    A --> B[B]

```

#### 3.3.2 Steps to Remove Unit Productions
The process of removing unit productions involves three main steps. Let's examine each step in detail with examples.
* Step 1. Find all unit pairs
* Step 2. Create new productions based on all unit pairs: for each unit pair (A,B), look at all non-unit productions with B on the left side, create new productions by replacing B with A
* Step 3. Remove original unit productions: identify all unit productions and remove them from the grammar; keep all new productions created in Step 2

### 3.4 Examples to Remove Unit Productions

#### Example 1: Remove unit productions from the given grammar
```
S → A
A → B
B → aB | b
```
Step 1: find all unit pairs: (S,A), (A,B), (S,B)
Step 2: create new productions:
* for (S,A), we do not find any non-unit productions with A on the left side
* for (A,B), we find two non-unit productions with B on the left side: B → aB | b
    - from B → aB, add new production A → aB
    - from B → b, add new production A → b
* for (S,B), we find two non-unit productions with B on the left side: B → aB | b
    - from B → aB, add new production S → aB
    - from B → b, add new production S → b
Step 3: remove all original unit productions: S → A and A → B should be removed.

After removing unit productions, the grammar is transformed to:
```
S → aB | b
A → aB | b
B → aB | b
```

#### Example 2: Remove unit productions from the given grammar
```
E → E + T | T
T → T * F | F
F → v | (E)
```
Step 1: find all unit pairs: (E,T), (T,F), (E,F)
Step 2: create new productions:
* for (E,T), we find non-unit productions with T on the left side: T → T * F
    - add new production E → T * F
* for (T,F), we find non-unit productions with F on the left side: F → v | (E)
    - from F → v, add new production T → v
    - from F → (E), add new production T → (E)
* for (E,F), we find non-unit productions with F on the left side: F → v | (E)
    - from F → v, add new production E → v
    - from F → (E), add new production E → (E)
Step 3: remove all original unit productions: E → T and T → F should be removed.

After removing unit productions, the grammar is transformed to:
```
E → E + T | T * F | v | (E)
T → T * F | v | (E)
F → v | (E)
```

### 3.5 Code Example: Removing Unit Productions

This function implements the three-step algorithm from Section 3.3.2. Before reading the code, recall the key concept from Section 3.3.1: a **unit pair**
(A, B) means that A can derive B through one or more unit productions. The full set of unit pairs is the **transitive closure** of the direct unit
production relation.

**How transitive closure is computed here:**

The `find_unit_pairs` inner function starts with reflexive pairs (A, A) for every non-terminal A — this represents the zero-step case "A derives A". It then iterates: if we know (X, Y) is a unit pair and we find a production Y → Z (a direct unit production), then X can derive Z, so we add (X, Z). This continues until no new pairs are found — the classic fixed-point loop
you saw in `find_nullable_symbols`.

**How new productions are built:**

After finding all unit pairs, the function creates a new grammar without any unit productions. For each non-unit production B → α and each unit pair (A, B), it adds A → α. This is Step 2 from Section 3.3.2: "inherit" all non-unit productions from B into A. Unit productions themselves are then discarded.

**Dependency:** This function calls neither `find_nullable_symbols` nor `remove_null_productions`. In the full CNF pipeline (Section 4.4), null productions are removed *before* unit productions.

**Outside Jupyter:** Copy this function plus the `Production`/`Grammar` classes into a `.py` file and run with `python unit_removal.py`.


````python
# ─────────────────────────────────────────────────────────────────────────────
# Step 2 of the transformation pipeline: Remove Unit Productions
#
# Implements Section 3.3.2 (three-step algorithm using transitive closure).
# Assumes null productions have already been removed (Section 2.6).
#
# Running outside Jupyter:
#   python unit_removal.py
# ─────────────────────────────────────────────────────────────────────────────


def remove_unit_productions(grammar):
    """
    Remove all unit productions A → B (Section 3.3.2).

    Strategy:
      1. Compute all unit pairs (A, B) via transitive closure (Section 3.3.1).
         A unit pair (A, B) means A ⟹* B through unit productions alone.
      2. For every unit pair (A, B) and every non-unit production B → α,
         add A → α to the new grammar (Section 3.3.2, Step 2).
      3. Discard all original unit productions (Step 3).

    Returns a new Grammar with no unit productions.
    """

    def find_unit_pairs():
        """
        Compute unit pairs via fixed-point iteration (Section 3.3.1).

        Start with reflexive pairs (A, A) — every non-terminal trivially
        "derives" itself in zero steps.
        Then, for each unit production X → Y and existing pair (Y, C),
        infer that X can also derive C in one more step, adding (X, C).
        Repeat until the set stabilizes (fixed point).
        """
        # Initialize with reflexive pairs: (A, A) for all non-terminals A
        pairs = {(A, A) for A in grammar.non_terminals}

        changed = True
        while changed:
            changed = False
            for prod in grammar.productions:
                # Only consider direct unit productions: single non-terminal RHS
                if len(prod.right) == 1 and prod.right[0] in grammar.non_terminals:
                    # prod encodes the direct unit production: prod.left → prod.right[0]
                    # If (prod.right[0], C) is already a known unit pair,
                    # then prod.left can also derive C → add (prod.left, C)
                    for B, C in list(pairs):
                        if B == prod.right[0] and (prod.left, C) not in pairs:
                            pairs.add((prod.left, C))
                            changed = True
        return pairs

    # Step 1: compute all unit pairs (transitive closure)
    unit_pairs = find_unit_pairs()

    # Steps 2 & 3: build new grammar
    new_productions = []
    for prod in grammar.productions:

        # Step 3: skip all original unit productions — they will be dropped
        if len(prod.right) == 1 and prod.right[0] in grammar.non_terminals:
            continue

        # Step 2: for every unit pair (A, B) where B == prod.left,
        # add A → prod.right (A inherits B's non-unit productions)
        for A, B in unit_pairs:
            if B == prod.left:
                new_productions.append(Production(A, prod.right))

    return Grammar(new_productions, grammar.start_symbol)


# ── Example: matches Section 3.4 Example 1 ────────────────────────────────────
# Original grammar: S → A; A → B; B → aB | b
# Unit pairs: (S,S), (A,A), (B,B), (S,A), (A,B), (S,B)
# After removal: S, A, and B all inherit B's non-unit productions → aB | b
grammar_with_unit = Grammar([
    Production('S', ['A']),         # unit production: S → A
    Production('A', ['B']),         # unit production: A → B
    Production('B', ['a', 'B']),    # non-unit production: B → aB
    Production('B', ['b']),         # non-unit production: B → b
], 'S')

print("Original grammar (Section 3.4 Example 1):")
print(grammar_with_unit)
print("\nGrammar after removing unit productions:")
print(remove_unit_productions(grammar_with_unit))
````

## 4. Chomsky Normal Form

### 4.1 What is Chomsky Normal Form?
Chomsky Normal Form (CNF) is a simplified form of context-free grammars where all productions follow specific patterns. A grammar is in CNF if and only if all productions are of the form:
```
A → BC (where A, B, and C are non-terminals)
A → a (where A is a non-terminal and a is a terminal)
```
### 4.2 What Makes CNF Valuable?
Chomsky Normal Form is particularly important for several reasons:
* It simplifies parsing algorithms (especially the CYK algorithm for membership testing)
* It makes certain proofs about context-free grammars easier, such as the Pumping Lemma for Context-free Languages
* It provides a standardized way to represent any context-free grammar
* It eliminates complex productions while preserving the generated language
* It is widely used in compiler design and natural language processing (NLP) for efficient parsing techniques.

### 4.3 Converting a CFG to CNF
To convert a given CFG into CNF, follow these steps:
* Step 1: Remove all Null Productions (previously covered in earlier sections)
* Step 2: Remove all Unit Productions (previously covered in earlier sections)
* Step 3: Convert Remaining Productions to CNF
    - Step 3a: Convert long right-hand sides into CNF: replace any production with three or more non-terminals on the right-hand size with new non-terminals, and repeat until all right-hand sides have exactly two non-terminals.
    - Step 3b: Convert terminals in mixed productions: replace  terminals in rules like A → aB  with a new non-terminal

#### Example 1: Converting a CFG to CNF
Conver the given CFG to CNF:
```
S → AB ∣ BCA ∣ aB
A → AB | a
B → b
C → c
```
* Step 1: No null productions
* Step 2: No unit productions
* Step 3a: the production S → BCA has three non-terminals on the right-hand side:
    - Introduce a new variable X such that: X → CA
    - Replace S → BCA with S → BX
* Step 3b: the production S → aB contains both a terminal (a) and a non-terminal (B). 
    - introduce a new non-terminal with a new production: Y → a
    - replace S → aB with S → YB

Final CNF grammar:
```
S → AB ∣ BX ∣ YB
A → AB | a
B → b
C → c
X → CA
Y → a
```
You may wonder, in step 3b, we introduced a new non-terminal Y with the production Y → a, instead of using the existing production A → a for the conversion. Here is why:
* If we had used A instead of Y, we would have replaced S → aB with S → AB. However, in the original grammar, A is already defined with its own role: A → AB | a, using A instead of Y could change the structure of the grammar, possibly affecting the language it generates. In this example, it could have introduced a new production S → ABB, which is not present in the original grammar.
* If we directly substituted A for a, it might create ambiguity in how non-terminals are structured. By introducing Y → a, we ensure that every production strictly adheres to CNF without modifying existing non-terminals.

Introducing Y ensures that the structure and meaning of the original CFG remain intact while following CNF's strict format. This is a standard transformation technique when converting CFGs to CNF. 

#### Example 2: Converting a CFG to CNF
Conver the given CFG to CNF:
```
S → aSb | bSa | SS | A
A → aAb | bAa | ∧
```
* Step 1: Remove all Null Productions
    - identify nullable non-terminals: A, S
    - from A → aAb | bAa, generate all combinations with nullable A: A → ab | ba
    - from S → aSb | bSa, generate all combinations with nullable S: S → ab | ba

Grammar becomes:
```
S → aSb | bSa | SS | A | ab | ba
A → aAb | bAa | ab | ba
```
* Step 2: Remove all unit productions:
    - Step 1. Find all unit pairs: (S,A)
    - Step 2. Create new productions for (S,A), look at all non-unit productions with A on the left side, create new productions by replacing A with S: S → aAb | bAa | ab | ba
    - Step 3. Remove original unit productions: keep all new productions created in Step 2

Grammar becomes:
```
S → aSb | bSa | SS | aAb | bAa | ab | ba
A → aAb | bAa | ab | ba
```
* Step 3: Convert Remaining Productions to CNF
    - Introduce two new non-terminals X and Y such that: X → a, Y → b
    - Grammar becomes:
        ```
        S → XSY | YSX | SS | XAY | YAX | XY | YX
        A → XAY | YAX | XY | YX
        X → a
        Y → b
        ```
    - Introduce two new non-terminals B, C, D, E such that: B → XS, C → YS, D → XA, E → YA
    - A long production such as S → XSY becomes S → BY, all other long productions can be converted to CNF

Final grammar becomes:
```
S → BY | CX | SS | DY | EX | XY | YX
A → DY | EX | XY | YX
X → a
Y → b
B → XS
C → YS
D → XA
E → YA
```

### 4.4 Code Example: Converting a CFG to Chomsky Normal Form

This function is the **culmination of the chapter**: it calls the two functions from Sections 2.6 and 3.5, then adds two more transformation steps to reach CNF. Recall from Section 4.3 that CNF requires every production to be exactly one of:

- **A → BC** (two non-terminals)
- **A → a**  (a single terminal)

The function performs four steps in order:

| Step | What it does | Where covered |
|---|---|---|
| 1 | Remove unit productions | Section 3.5 |
| 2 | Remove null productions | Section 2.6 |
| 3a | Replace terminals in mixed productions | Section 4.3 Step 3b |
| 3b | Break long RHS (≥ 3 symbols) into binary pairs | Section 4.3 Step 3a |

**Important: Steps 1 and 2 are called in the order shown in the code, which differs from the order listed in Section 4.3.** The code removes unit productions *before* null productions. Both orderings are valid — null removal may create new unit productions but unit removal does not create new null productions, so removing unit productions first is safe and often produces
a slightly cleaner result.

**How Step 3a works (terminal wrapping):**
A mixed production like `A → aB` violates CNF because it contains both a terminal (`a`) and a non-terminal (`B`). The fix is to introduce a fresh non-terminal `T_a` with `T_a → a` and rewrite `A → aB` as `A → T_aB`. The `terminal_rules` dictionary caches these mappings so the same terminal never gets two different wrapper non-terminals.

**How Step 3b works (binary splitting):**
A long production like `S → ABC` violates CNF because it has three symbols on the RHS. The fix is to peel off the *rightmost* two symbols, replace them with a new non-terminal, and repeat. For `S → ABC`:
- Create `X_n → BC`, giving `S → AX_n` ✓

The `while len(new_right) > 2` loop does this peeling iteratively. The auto-generated names `X_0`, `X_6`, etc. use the current production count as a uniqueness stamp — ugly but guaranteed collision-free.

**Outside Jupyter:** Copy all three functions (`remove_null_productions`, `remove_unit_productions`, `to_chomsky_normal_form`) plus the `Production`/`Grammar` classes into `cnf_converter.py` and run with: 
```bash
python cnf_converter.py
```


````python
# ─────────────────────────────────────────────────────────────────────────────
# Step 3: Convert to Chomsky Normal Form (CNF)
#
# Implements Section 4.3. Calls remove_unit_productions and
# remove_null_productions (defined in earlier cells) as sub-steps.
#
# CNF requires every production to be either:
#   A → BC  (exactly two non-terminals)
#   A → a   (exactly one terminal)
#
# Running outside Jupyter:
#   python cnf_converter.py
#   (Include all earlier class/function definitions in the same file)
# ─────────────────────────────────────────────────────────────────────────────


def to_chomsky_normal_form(grammar):
    """
    Convert a CFG to Chomsky Normal Form (Section 4.3).

    Applies four transformations in sequence:
      1. Remove unit productions (Section 3.5)
      2. Remove null productions (Section 2.6)
      3a. Wrap terminals in mixed productions with new non-terminals (Step 3b)
      3b. Break productions with 3+ RHS symbols into binary pairs (Step 3a)

    Returns a new Grammar in CNF.
    """

    # ── Sub-steps 1 & 2: eliminate unit and null productions ─────────────────
    # These are called first so that Steps 3a/3b only ever see productions that
    # are already free of unit and null issues.
    grammar = remove_unit_productions(grammar)  # Section 3.5
    grammar = remove_null_productions(grammar)  # Section 2.6

    # ── Sub-steps 3a & 3b: enforce CNF structure ─────────────────────────────
    new_productions = []
    # Maps each terminal symbol to its dedicated wrapper non-terminal
    # e.g., 'a' → 'T_a', so we never create two non-terminals for the same terminal
    terminal_rules = {}

    for prod in grammar.productions:

        # Productions already in CNF: A → a (single terminal)
        # Pass through unchanged.
        if len(prod.right) == 1 and prod.right[0] in grammar.terminals:
            new_productions.append(prod)
            continue

        # ── Step 3a: wrap terminals in mixed/long productions ─────────────────
        # A → aB violates CNF; replace 'a' with T_a and add T_a → a.
        # A → a  is already CNF (handled above); skip it here.
        new_right = []
        for symbol in prod.right:
            if symbol in grammar.terminals:
                # Introduce a wrapper non-terminal for this terminal if needed
                if symbol not in terminal_rules:
                    wrapper = f"T_{symbol}"          # e.g., T_a for terminal 'a'
                    terminal_rules[symbol] = wrapper
                    new_productions.append(Production(wrapper, [symbol]))
                new_right.append(terminal_rules[symbol])  # replace 'a' with T_a
            else:
                new_right.append(symbol)  # non-terminals pass through unchanged

        # ── Step 3b: break long RHS into binary pairs ─────────────────────────
        # A → X₁X₂X₃ becomes: X_n → X₂X₃, A → X₁X_n
        # Peel off rightmost two symbols repeatedly until exactly two remain.
        while len(new_right) > 2:
            # Use production count as a unique ID for the new non-terminal
            new_nt = f"X_{len(new_productions)}"
            new_productions.append(Production(new_nt, new_right[-2:]))  # X_n → last two
            new_right = new_right[:-2] + [new_nt]                       # replace with X_n

        # Add the (now binary) production
        new_productions.append(Production(prod.left, new_right))

    return Grammar(new_productions, grammar.start_symbol)


# ── Example: matches Section 4.3 Example 1 ────────────────────────────────────
# Original: S → ABC | aB; A → AB | a; B → b; C → c
# No null or unit productions, so Steps 1 & 2 are no-ops here.
# Step 3a: wrap 'a' in S → aB → introduce Y → a (stored as T_a)
# Step 3b: break S → ABC (3 symbols) → introduce X → CA, rewrite S → BX
non_cnf_grammar = Grammar([
    Production('S', ['A', 'B', 'C']),
    Production('A', ['a', 'A']),
    Production('B', ['b']),
    Production('C', ['c', 'A', 'B']),
], 'S')

print("Original grammar (Section 4.3 Example 1 variant):")
print(non_cnf_grammar)
print("\nGrammar in Chomsky Normal Form:")
print(to_chomsky_normal_form(non_cnf_grammar))
````
